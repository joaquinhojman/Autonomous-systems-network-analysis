```python
import networkx as nx
import matplotlib.pyplot as plt
import scipy as sp
import numpy as np
import matplotlib.cm as cm
```

```python
G = nx.read_edgelist("as.txt", delimiter="	")
```

```python
print("Is directed? " + str(nx.is_directed(G)))
```

```python
from matplotlib.pyplot import figure

def print_graph(G, pos): 
    figure(figsize=(12, 12), dpi=90)
    # nodes
    nx.draw_networkx_nodes(G, pos, node_size=8)
    # edges
    nx.draw_networkx_edges(G, pos, width=0.5,alpha=0.2)
    plt.axis('off')
    plt.figure(figsize=(50000,50000))
    plt.show()
```

```python
pos_spring_layout = nx.spring_layout(G)
print_graph(G, pos_spring_layout)
```

```python
print("Nodes length: "+str(len(G.nodes)))
print("Edges length: "+str(len(G.edges)))
```

```python
print("Diameter: " + str(nx.diameter(G)))
```

```python
degrees = G.degree()
degree = 0
for i in range(0,len(G.nodes)):
    degree += degrees[str(i)]
print("Degree: " + str(degree))
print("Average degree: " + str(degree/len(G.nodes)))
```

```python
from operator import itemgetter
max_connected = max(degrees, key=itemgetter(1))
print("More connected node: " + str(max_connected[0]) + " with: " + str(max_connected[1]) + " edges")
```

```python
print("Average_clustering: " + str(nx.average_clustering(G)))
```

```python
print("Global bridges: " + str(len(set(nx.bridges(G)))))
```

```python
print("Local bridges: " + str(len(set(nx.local_bridges(G)))))
```

```python
print("Articulation points: " + str(len(set(nx.articulation_points(G)))))
```

```python
print("Graph is connected?: " + str(nx.is_connected(G)))
```

```python
print("Number of self loops: " + str(nx.number_of_selfloops(G)))
```
